```python
BKK = pd.read_csv('BKK.csv')
NON = pd.read_csv('Nonthaburi.csv')
PT = pd.read_csv('PathumThani.csv')
SP = pd.read_csv('SamutPrakan.csv')
SS = pd.read_csv('SamutSakhon.csv')
```

```python
NON = NON[[' PM2.5']]
NON
```

```python
NON.rename(columns= {' PM2.5' : 'NON2.5'}, inplace = True)
NON
```

```python
PT = PT[[' PM2.5']]
PT
```

```python
PT.rename(columns= {' PM2.5' : 'PT2.5'}, inplace = True)
PT
```

```python
SP= SP[[' PM2.5']]
SP
```

```python
SP.rename(columns= {' PM2.5' : 'SP2.5'}, inplace = True)
SP
```

```python
SS = SS[[' PM2.5']]
SS
```

```python
SS.rename(columns= {' PM2.5' : 'SS2.5'}, inplace = True)
SS
```

```python
frames1 = pd.concat([BKK,NON,PT,SP,SS], axis=1 ,join = 'outer')
frames1
```

```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
frames1.corr()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.heatmap(frames1.corr(), annot = True)
```
