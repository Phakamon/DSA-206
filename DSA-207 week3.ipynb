```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Fdata = pd.read_csv('fishdata.csv', encoding = 'unicode_escape')
```

```python
Fdata.head(5)
```

```python
df = Fdata.drop(labels=0, axis = 0)
df.head(5)
```

```python
df1 = df.groupby(['Month'])['Carbon Monoxide','External temperature','Humidity'].mean().reset_index()
```

```python
df1.set_index("Month", inplace = True)
```

```python
df1
```

```python
df1['Humidity levels'] = pd.qcut(df['Humidity'], q = 3)
df1
```

```python
df1['Humidity levels'].value_counts()
```

```python
Humidity_in = ['Low','Medium','High']
df1['Humidity Rank'] = pd.qcut(df['Humidity'], q = 3, labels= Humidity_in)
df1
```

```python
df1['Humidity Rank'].value_counts()
```
